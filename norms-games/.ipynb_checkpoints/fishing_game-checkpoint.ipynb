# Ostrom's Fishing Game in Extensive Form

See page 85, Chapter 4 in E. Ostrom, R. Gardner, J. Walker, Rules, Games, and Common-Pool Resources, The University of Michigan Press, Ann Arbor, 1994. https://doi.org/10.3998/mpub.9739.

The utilities at the terminal nodes are parametrized upon:

* $v_i,\; i=1,2$: economical value of the fishing spots. Spot 1 is more valuable than spot 2: $v_1>v_2$.

* $P$: probability that fisher 1 wins a fight if one occurs. Fisher 1 is the strongest player, $P>0.5$.

* $c$: cost of travel between the two spots. This cost is smaller than the value of the less valuable spot: $c<v_2$.

* $d$: damage incurred when a fight is lost.


Reproduce the game, start by assuming perfect information, no ``chance`` nodes.

```python
from extensivegames import ExtensiveFormGame, plot_game, backward_induction, subgame_perfect_equilibrium_path
```

```python
fishing_game = ExtensiveFormGame(title='Fishing game')

# add the two positions for the two fishers
fishing_game.add_positions(['fisher 1', 'fisher 2'])

# add the nodes to the graph
fishing_game.add_node(1, position_turn='fisher 1', is_root=True)
fishing_game.add_node(2, position_turn='fisher 2')
fishing_game.add_node(3, position_turn='fisher 2')
fishing_game.add_node(4, position_turn='fisher 1')
fishing_game.add_node(5)
fishing_game.add_node(6)
fishing_game.add_node(7, position_turn='fisher 1')
fishing_game.add_node(8, position_turn='fisher 2')
fishing_game.add_node(9, position_turn='fisher 2')
fishing_game.add_node(10, position_turn='fisher 2')
fishing_game.add_node(11, position_turn='fisher 2')
for i in range(12, 20):
    fishing_game.add_node(i)

# add the edges to the graph
fishing_game.add_edge(1, 2, label='go to spot 1')
fishing_game.add_edge(1, 3, label='go to spot 2')
fishing_game.add_edge(2, 4, label='go to spot 1')
fishing_game.add_edge(2, 5, label='go to spot 2')
fishing_game.add_edge(3, 6, label='go to spot 1')
fishing_game.add_edge(3, 7, label='go to spot 2')
fishing_game.add_edge(4, 8, label='stay')
fishing_game.add_edge(4, 9, label='leave')
fishing_game.add_edge(7, 10, label='stay')
fishing_game.add_edge(7, 11, label='leave')
fishing_game.add_edge(8, 12, label='stay')
fishing_game.add_edge(8, 13, label='leave')
fishing_game.add_edge(9, 14, label='stay')
fishing_game.add_edge(9, 15, label='leave')
fishing_game.add_edge(10, 16, label='stay')
fishing_game.add_edge(10, 17, label='leave')
fishing_game.add_edge(11, 18, label='stay')
fishing_game.add_edge(11, 19, label='leave')

# add imperfect information, equivalent to having players take the actions simultaneously
fishing_game.set_information_partition('fisher 2', [{2,3}, {8,9}, {10,11}])

  
# utility parameters
v1, v2 = 5, 3
P = 0.6
c = 0.5
d = 2

def set_utilities(v1: float, v2: float, P: float, c: float, d: float, game: ExtensiveFormGame):
    w11 = P*v1+(1-P)*(-d)
    w12 = P*v2+(1-P)*(-d)
    w21 = (1-P)*v1+P*(-d)
    w22 = (1-P)*v2+P*(-d)
    # utilities (in an economical sense) to the terminal nodes
    game.set_utility(5, {'fisher 1':v1, 'fisher 2':v2})
    game.set_utility(6, {'fisher 1':v2, 'fisher 2':v1})
    game.set_utility(12, {'fisher 1':w11, 'fisher 2':w21})
    game.set_utility(13, {'fisher 1':v1, 'fisher 2':v2-c})
    game.set_utility(14, {'fisher 1':v2-c, 'fisher 2':v1})
    game.set_utility(15, {'fisher 1':w12-c, 'fisher 2':w22-c})
    game.set_utility(16, {'fisher 1':w12, 'fisher 2':w22})
    game.set_utility(17, {'fisher 1':v2, 'fisher 2':v1-c})
    game.set_utility(18, {'fisher 1':v1-c, 'fisher 2':v2})
    game.set_utility(19, {'fisher 1':w11-c, 'fisher 2':w21-c})
    
set_utilities(v1, v2, P, c, d, fishing_game)
```

```python
# default keywords for rendering the figure
my_fig_kwargs = dict(figsize=(40,30), frameon=False)
my_node_kwargs = dict(font_size=38, node_size=5000, edgecolors='k', linewidths=3)
my_edge_kwargs = dict(arrowsize=50, width=5)
my_edge_labels_kwargs = dict(font_size=28)
my_patch_kwargs = dict(linewidth=3)
my_legend_kwargs = dict(fontsize=38, loc='upper right', edgecolor='white')
my_utility_label_kwargs = dict(horizontalalignment='center', fontsize=30)
my_info_sets_kwargs = dict(linestyle='--', linewidth=3)

position_colors =  {'fisher 1':'aquamarine', 'fisher 2':'greenyellow'}

fig = plot_game(fishing_game,
                position_colors,
                fig_kwargs=my_fig_kwargs,
                node_kwargs=my_node_kwargs,
                edge_kwargs=my_edge_kwargs,
                edge_labels_kwargs=my_edge_labels_kwargs,
                patch_kwargs=my_patch_kwargs,
                legend_kwargs=my_legend_kwargs,
                utility_label_kwargs=my_utility_label_kwargs,
                utility_label_shift=0.06,
                info_sets_kwargs=my_info_sets_kwargs)
```

Note that the utilities here are arbitrary. We compute the value assigned by every position to every node with **backward induction**, from which we obtain the **subgame perfect equilibrium** path from the root node to a terminal node.

```python
node_values = backward_induction(fishing_game, 1)
for n, v in node_values.items():
    print("{:>2}: ".format(n), end="")
    for p, u in v.items():
        print("{:>10}: {:.2f}".format(p, u), end="")
    print()
```

```python
path = subgame_perfect_equilibrium_path(fishing_game, 1)
print(path)
```

## Default configuration rule



**Case 1**: Both agents are similarly likely to win the fight ($P\sim 0.5$), and spot is much better than spot 2 ($v_1>>v_2$).

```python
v1, v2 = 10, 1
P = 0.5
c = 0.5
d = 2

w11 = P*v1+(1-P)*(-d)
w12 = P*v2+(1-P)*(-d)
w21 = (1-P)*v1+P*(-d)
w22 = (1-P)*v2+P*(-d)

print(w11, w12)
print(w21, w22)

set_utilities(v1, v2, P, c, d, fishing_game)
```
